#2 What is an interpreter in Python ?
   A Python interpreter is a program that reads and executes Python code line by line, translating it into machine instructions instantly. It enables running Python scripts without compiling them first, making coding and debugging fast and interactive.
#3 What are pre-defined keywords in Python ?

#4 Can keywords be used as variable names ?

#5 What is mutability in Python ?
   Mutability in Python refers to an object's ability to have its value or state changed after it has been created. Mutable objects, like lists and dictionaries, can be altered without changing their identity, whereas immutable objects, like strings and tuples, cannot be changed once created.

#6 Why are lists mutable, but tuples are immutable ?
   Lists are mutable because they are designed to allow modifications such as adding, removing, or changing elements after creation. Tuples are immutable to provide a fixed, unchangeable sequence which is useful for protecting data.

#9 What is type casting in Python?
   Type casting in Python is the process of converting a variable from one data type to another. it can done  explicitly or implicitly as needed
#10 What is the difference between implicit and explicit type casting?
   The difference between implicit and explicit type casting is that implicit type casting happens automatically by the compiler when converting data types, while explicit type casting requires to manually specify the conversion. Implicit casting is automatic and convenient, whereas explicit casting gives more control but requires clear instructions from the programmer.
#11 What is the purpose of conditional statements in Python?
   The purpose of conditional statements in Python is to make decisions in a program by executing certain blocks of code only if specific conditions are true or false. They control the flow of the program, allowing it to behave differently in different situations based on those conditions.
#12 How does the elif statement work?
   The elif statement in Python checks another condition if the previous if statement is false. It allows multiple conditions to be tested sequentially until one is true, then executes the corresponding code block
#13 What is the difference between for and while loops?
   The difference between for and while loops is:
   A for loop is used when the number of iterations is known in advance, iterating over a sequence or a range.
   A while loop runs as long as a given condition remains true, and is used when the number of iterations is not   predetermined.
#14 Describe a scenario where a while loop is more suitable than a for loop?
   A while loop is more suitable than a for loop when the number of iterations is unknown and the loop needs to continue based on a condition, such as waiting for user input or reading data until a certain event occurs. In contrast, a for loop is ideal when the number of iterations is predetermined or fixed, like iterating over a range or a collection of items

```python
#6  Write a function to demonstrate the behavior of mutable and immutable arguments?


def demonstrate_mutable_immutable(num, num_list):
    print("Inside function before change:")
    print("num =", num)
    print("num_list =", num_list)

    # Trying to modify immutable argument
    num += 10   # This creates a new integer object

    # Modifying mutable argument
    num_list.append(100)  # This modifies the existing list

    print("Inside function after change:")
    print("num =", num)
    print("num_list =", num_list)


# Driver code
x = 5                # Immutable (integer)
y = [1, 2, 3]        # Mutable (list)

print("Before calling function:")
print("x =", x)
print("y =", y)

demonstrate_mutable_immutable(x, y)

print("After calling function:")
print("x =", x)   # unchanged (immutable)
print("y =", y)   # changed (mutable)
```

```python
#7   Write a program that performs basic arithmetic operations on two user-input numbers?

# Basic Arithmetic Operations

# Taking input from the user
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Performing operations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2

# Handling division by zero
if num2 != 0:
    division = num1 / num2
else:
    division = "Undefined (division by zero)"

# Displaying results
print("\nResults:")
print(f"{num1} + {num2} = {addition}")
print(f"{num1} - {num2} = {subtraction}")
print(f"{num1} * {num2} = {multiplication}")
print(f"{num1} / {num2} = {division}")
```

```python
#8  Write a program to demonstrate the use of logical operators?


# Program to demonstrate logical operators

# Taking two numbers
a = 10
b = 5

print("The values are: a =", a, "and b =", b)

# Using 'and' operator
if a > 0 and b > 0:
    print("Both a and b are positive numbers (AND condition is True)")

# Using 'or' operator
if a > 0 or b < 0:
    print("At least one condition is True (OR condition is True)")

# Using 'not' operator
if not (a < 0):
    print("a is not a negative number (NOT condition is True)")
```

```python
#9 Write a Python program to convert user input from string to integer, float, and boolean types?


# Program to convert user input into integer, float, and boolean types
user_input = input("Enter a value: ")
try:
    int_value = int(user_input)
    print("Integer value:", int_value)
except ValueError:
    print("Cannot convert to integer.")
try:
    float_value = float(user_input)
    print("Float value:", float_value)
except ValueError:
    print("Cannot convert to float.")

if user_input.lower() in ["true", "1", "yes"]:
    bool_value = True
elif user_input.lower() in ["false", "0", "no"]:
    bool_value = False
else:
    bool_value = bool(user_input)

print("Boolean value:", bool_value)
```

```python
#10  Write code to demonstrate type casting with list elements?



# List with string numbers
string_numbers = ["10", "20", "30", "40", "50"]

# Casting elements from string to int
int_numbers = [int(num) for num in string_numbers]
print("String to Int:", int_numbers)

# Casting elements from int to float
float_numbers = [float(num) for num in int_numbers]
print("Int to Float:", float_numbers)

# Casting elements from float to string
string_again = [str(num) for num in float_numbers]
print("Float to String:", string_again)

# Casting integers to boolean
bool_values = [bool(num) for num in int_numbers]
print("Int to Boolean:", bool_values)

# Mixed type list casting to string for uniformity
mixed_list = [123, 45.6, True, None]
string_casted = [str(item) for item in mixed_list]
print("Mixed to String:", string_casted)
```

```python
#13   Write a Python program to find the sum of all even numbers between 1 and 50?


# Program to find the sum of all even numbers between 1 and 50

total = 0

for num in range(1, 51):
    if num % 2 == 0:
        total += num

print("Sum of even numbers between 1 and 50 is:", total)
```

```python
#14 Write a program to reverse a string using a while loop?


# Input string
text = "Hello World"

# Initialize variables
reversed_text = ""
index = len(text) - 1

# Loop to reverse the string
while index >= 0:
    reversed_text += text[index]
    index -= 1

# Print the reversed string
print("Original String:", text)
print("Reversed String:", reversed_text)
```

```python
#15 . Write a Python program to calculate the factorial of a number provided by the user using a while loop?


# Factorial calculation using a while loop

# Take input from user
num = int(input("Enter a number: "))

# Initialize variables
factorial = 1
i = 1

# Use while loop to calculate factorial
while i <= num:
    factorial *= i
    i += 1

print(f"The factorial of {num} is {factorial}")
```
